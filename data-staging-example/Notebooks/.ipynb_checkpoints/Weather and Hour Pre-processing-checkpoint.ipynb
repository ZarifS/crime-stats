## Steps to find the nearest weather stations (Not needed for the demo)

```python
longitude = coll_2017_df['Longitude']
latitude = coll_2017_df['Latitude']
```

```python
stations_df = pd.read_csv('../Data/Station Inventory EN.csv')
stations_df = stations_df[stations_df['Province'] == 'ONTARIO']
```

```python
station_long = list(stations_df['Longitude (Decimal Degrees)'])
station_lat = list(stations_df['Latitude (Decimal Degrees)'])
station_name = list(stations_df['Name'])
```

```python
near_station = []

for index, col_long in enumerate(longitude):
    distance = []
    
    for idx, long in enumerate(station_long):
        coords_1 = (latitude[index], longitude[index])
        coords_2 = (station_lat[idx], station_long[idx])

        distance.append(geopy.distance.distance(coords_1, coords_2).km)

    near_station.append([station_name[distance.index(x)] for x in sorted(distance)[:100]])
    break
```

```python
near_stations_df = pd.DataFrame(near_station, columns=['Name'])
coll_2017_df_new = pd.concat([coll_2017_df, near_stations_df], axis=1)
```
