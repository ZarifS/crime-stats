```python
import pandas as pd
denver_parent_data = pd.read_csv("../datasets/denver-data.csv") 
```

```python
denver_address = pd.read_csv("./denver-address.csv")
denver_crime = pd.read_csv("./denver-crime.csv")
denver_date = pd.read_csv("../merged-datasets/merged-dates.csv")
```

```python
#Adding Denver crime-key
denver_parent_data = pd.concat([denver_parent_data, denver_crime], axis=1)
denver_parent_data.head()
```

```python
#only picking colummns that are needed
denver_parent_data = denver_parent_data[['crime_key', 'INCIDENT_ADDRESS', 'FIRST_OCCURRENCE_DATE', 'IS_TRAFFIC', 'crime_report_time', 'crime_type']]
```

```python
#Adding Denver address-key
denver_parent_data = pd.merge(denver_parent_data, denver_address, how='left', left_on=['INCIDENT_ADDRESS'], right_on = ['location_name'])
```

```python
denver_parent_data.head()
```

```python
#Drop unnecessary columns
denver_parent_data = denver_parent_data.iloc[:, 0:7]
denver_parent_data.head()
```

```python
#Split Reported time into year, month, day
date_values = denver_parent_data['FIRST_OCCURRENCE_DATE'].values

year = []
month = []
day = []
completeDay = []

for date in date_values:
    date = date.split() # Split on whitespace
    date = date[0] # Grab first portion which m/d/y
    completeDay.append(date)# Store the complete day to remove duplicates later
    date = date.split('/') # Split into m, d and y and turn into int
    month.append(int(date[0]))
    day.append(int(date[1]))
    year.append(int(date[2]))

denver_parent_data['year'] = year
denver_parent_data['month'] = month
denver_parent_data['day'] = day
denver_parent_data.head()
```

```python
#Add Date Key
denver_parent_data = pd.merge(denver_parent_data, denver_date, how='left', left_on=['year', 'month', 'day'], right_on = ['year', 'month', 'day'])
```

```python
#Create denver_fact_table
denver_fact_table = denver_parent_data[['location_key', 'date_key', 'crime_key', 'IS_TRAFFIC', 'crime_report_time', 'crime_type']]
denver_fact_table.head()
```

```python
# Add a new column for is_traffic and is_fatal
is_fatal = []
for crime_type in denver_fact_table['crime_type']:
    if crime_type == 'Homicide' or crime_type == 'Vehicle Collision or Pedestrian Struck (with Fatality)':
        is_fatal.append(1)
    else: is_fatal.append(0)

denver_fact_table['is_fatal'] = is_fatal
denver_fact_table.head()
```

```python
def is_nighttime(row):
    hrs = row['crime_report_time'][0].split(':')[0]
    if int(hrs) >= 18 or int(hrs) <= 6:
        return 1
    else:
        return 0

denver_fact_table['is_nighttime'] = denver_fact_table.apply (lambda row: is_nighttime(row), axis=1)
```

```python
denver_fact_table.head()
```

```python
denver_fact_table.drop(['crime_type', 'crime_report_time'], axis=1, inplace=True)
denver_fact_table.head()
```

```python
denver_fact_table = denver_fact_table.rename(columns={"IS_TRAFFIC": "is_traffic"})
denver_fact_table.head()
```

```python
denver_fact_table_cols = denver_fact_table.columns.tolist()
denver_table_new_cols = denver_fact_table_cols[0:3] + [denver_fact_table_cols[5]] + [denver_fact_table_cols[4]] + [denver_fact_table_cols[3]]

denver_fact_table = denver_fact_table[denver_table_new_cols]

denver_fact_table.head()
```

```python
denver_fact_table.to_csv('./denver_fact_table.csv', index=False)
```
